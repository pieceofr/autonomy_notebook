## Data-set from people in MSA  
The Autonomy sefrom today and yesterdaylect **today** and **yesterday** last report of each user inside the **MSA**. The Data-set consist with
+ total count of customized symptom report
+ total people in the data-set

```python
totalCustomizedCount = 5
totalPeople = 10
```

+ total count of each official symptom in the Data-set (we call it distribution)

```python

symptomDistribution = {
    'Fever':   4.0,
    'Cough':   3.0,
    'Fatigue': 1.0,
    'Breath':  1.0,
    'Nasal':   1.0,
    'Throat':  2.0,
    'Chest':   0.0,
    'Face':    0.0,
}
```

+ weight metrix : user can define the weight for each official symptoms while customized weight always weight 1

```python

symptomWeightMatrix = {
    'Fever':   3,
    'Cough':   2,
    'Fatigue': 1,
    'Breath':  1,
    'Nasal':   1,
    'Throat':  1,
    'Chest':   2,
    'Face':    2,
}
```

```python
totalWeight = 0.0
for key, value  in symptomDistribution.items():
    totalWeight = totalWeight + value*symptomWeightMatrix[key]

print('totalWeight',totalWeight)
maxWeightPerPerson = 0.0
for key, value  in symptomWeightMatrix.items():
    maxWeightPerPerson = maxWeightPerPerson + value
print('maxWeightPerPerson:',maxWeightPerPerson)
```

#### Final Score

```python
if ((totalPeople*maxWeightPerPerson)+totalCustomizedCount*1) > 0:
    score = 100*(1-(totalWeight/((totalPeople*maxWeightPerPerson)+totalCustomizedCount*1)))
else:
    score = 100
print('score', score)
```
